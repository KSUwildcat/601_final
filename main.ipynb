```python
exports = pd.read_excel('statista/electricity-exports-from-the-us.xlsx')
```

```python
exports
```

```python
X = np.arange(1, 14)
X = X.reshape(-1, 1)
Y = exports.iloc[:, 1].values.reshape(-1, 1)
linear_regressor = LinearRegression()
linear_regressor.fit(X, Y)
Y_pred = linear_regressor.predict(X)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.title('US Electricity Exports (Linear Regression)')
plt.xlabel('Year Progression')
plt.ylabel('Exports')
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X, Y, color='red')
plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('US Electricity Exports (2nd degree Regression)')
plt.xlabel('Year Progression')
plt.ylabel('Exports')
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X, Y, color='red')
plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('US Electricity Exports (3rd degree Regression)')
plt.xlabel('Years progression')
plt.ylabel('Exports')
plt.show()
```

```python
cell_cost = pd.read_excel('iea/us-quarterly-prices-of-photovoltaic-multi-modules.xlsx')
```

```python
cell_cost
```

```python
X = cell_cost.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
X = np.arange(1, 19)
X = X.reshape(-1, 1)
Y = cell_cost.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=1)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly, Y)

plt.figure(figsize=(20,10))
plt.scatter(X, Y, color='red')
plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
